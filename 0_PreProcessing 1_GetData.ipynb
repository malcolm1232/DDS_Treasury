```python
#Minmax Scaler
def get_minmax_scaled(df):
    columns_that_are_numerical = df.dtypes != object
    
    columns_that_are_numerical = df.columns[columns_that_are_numerical].tolist()
    print('columns_that_are_numerical:',columns_that_are_numerical)

    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    print('df(before minmax):\n',df)
    df[columns_that_are_numerical] = scaler.fit_transform(df[columns_that_are_numerical])
    print('df(aft minmax):\n',df)
    return df
```

```python
def get_LabelEncoded(df):
#     from sklearn.preprocessing import MinMaxScaler

    from sklearn.preprocessing import LabelEncoder
    categorical_feature_mask = df.dtypes==object
    print('categorical_feature_mask:',categorical_feature_mask)
    categorical_cols = df.columns[categorical_feature_mask].tolist()
    print('categorical_cols:',categorical_cols)
    col_loc_list =  []
    for i in categorical_cols:
        loc = df.columns.get_loc(i)
        col_loc_list.append(loc)
    print('col_loc_list:',col_loc_list)

    # __________________________________________________________________________________________________#Select FLoat and INT
    categorical_feature_mask_float_int = df.dtypes!=object
    print('categorical_feature_mask_float_int:',categorical_feature_mask_float_int)
    categorical_feature_mask_float_int = df.columns[categorical_feature_mask_float_int].tolist()
    print('categorical_feature_mask_float_int:',categorical_feature_mask_float_int)
    categorical_feature_mask_float_int_list =  []
    for i in categorical_feature_mask_float_int:
        loc = df.columns.get_loc(i)
        categorical_feature_mask_float_int_list.append(loc)
    print('categorical_feature_mask_float_int_list:',categorical_feature_mask_float_int_list)


    print('categorical_feature_mask_float_int_list_to_list1:',df.columns.tolist())
    print('df.dtypes2:',df.dtypes)

    # _________________________________________________________________________________________________________# Commence Label Encode
    encoder_le = LabelEncoder()
#     encoder_MinMax = MinMaxScaler()
    print('col_loc_list (before Label Encoding OBJECTs):',col_loc_list)


    # __________________________________________________________________________________________________________#Label Encoding
    df_col_loc_list = df.loc[:,categorical_cols] #obtain categorical columns
    print('df_col_loc_list:',df_col_loc_list)
    df_col_loc_list.fillna('FillSC', inplace= True)
    le = LabelEncoder()
    df_encoded_SCType0123 = df_col_loc_list.apply(le.fit_transform) #label encode categorical columns
    print('df_encoded_SCType0123:\n',df_encoded_SCType0123)
    df.drop(categorical_cols, axis =1 , inplace = True) #drop SCType
    print('df_line80:\n',df)
    df = pd.concat([df,df_encoded_SCType0123], axis =1) #check sort
    print('df_line83:\n',df)

    return df
```
