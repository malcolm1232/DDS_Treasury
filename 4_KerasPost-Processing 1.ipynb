# Putting it all together

```python
import pandas as pd
csv_fname = 'capital-gain'#marital-status_1
fname_category = 'capital-gain' # column name #marital-status
fname_category_code = 'capitalgain'# object # marital_status
exec(f'threshold_{fname_category_code} = []')
exec(f'accuracy_count_store_{fname_category_code} = []')

res = pd.read_csv('./results/DBS_Part4_Type2_{}.csv'.format(csv_fname),index_col = 0)
df_test_scaled = pd.read_csv('./dataset/adult_test_RenameCol_MinMax_LE.csv',index_col = 0)
df_test_orig = pd.read_csv('./dataset/adult_test.csv',index_col = 0)

df_col = df_test_scaled.columns.to_list()
res_col = res.columns.to_list()
print("res_col:",res_col)
desired_col = fname_category

print("desired_col:",desired_col)
df_test_orig_to_concat_colnum = df_test_orig.loc[:, desired_col]
# df_test_orig_to_concat_colnum.columns = ['Orig_{}'.format(desired_col)]

df_test_scaled_to_concat_colnum = df_test_scaled.loc[:, desired_col]
# df_test_scaled_to_concat_colnum.columns = 'Scaled_{}'.format(desired_col)
df_final = pd.concat([df_test_orig_to_concat_colnum, df_test_scaled_to_concat_colnum, res],axis=1)
colname1 = desired_col + '_Actual'
colname2 = desired_col + '_Scaled'
df_final.columns = [ colname1, colname2, res_col[0], res_col[1], res_col[2]]

df_final.to_csv('./results/DBS_Part4_Type2_{}_processed.csv'.format(csv_fname ))

df = df_final.copy()
print("df:\n",df)

threshold_list = [0.1, .2, 0.2094,0.22, .25, .3,.35, .4,.45, .5, .6, .7]

for threshold in threshold_list:

    csv_colName = 'P_' + fname_category
    print("csv_colName:",csv_colName)
    P_list = df.loc[:, csv_colName]
    binary_list = []
    class_list = df['class'].to_list()
#     print("P_list:",P_list)
    for p in P_list:
        if p >= threshold:
            binary_list.append(1)
        else:
            binary_list.append(0)

    count = 0
    #count number of correct ONES
    for i in range(len(df)):
        if class_list[i] == 1:
            if binary_list[i] == 1 :
                count+=1
    
    #Count overall accuracy 
    accuracy_count = 0
    for i in range(len(df)):
        if binary_list[i] == class_list[i]:
            accuracy_count+=1
            
    storing_thresh = 'Threshold: {} Correct: {}, percentage: {}'.format(threshold, count, count/len(df))
    exec(f'threshold_{fname_category_code}.append(storing_thresh)')
    accuracy_count_store = accuracy_count / len(df)
    exec(f'accuracy_count_store_{fname_category_code}.append(storing_thresh)')

    print('Threshold: {} Correct: {}, percentage: {}'.format(threshold, count, count/len(df)))
    print("accuracy_count:",accuracy_count / len(df))
    print('')
```
