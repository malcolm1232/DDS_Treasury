```python
import pandas as pd
dfy = pd.read_csv('./dataset/adult_data_RenameCol_MinMax_LE.csv',index_col = 0) #adult_data_RenameCol_MinMax_LE
dfx = pd.read_csv('./dataset/adult_data.csv',index_col = 0) #adult_data_RenameCol_MinMax_LE
dfx['class'] = dfy['class']
df = dfx
df
```

```python
df_0 = df[df['class'] == 0]
df_1 = df[df['class'] == 1]
df_1
```

```python
df_0_sam = df_0.sample(frac=1)
df_1_sam = df_1.sample(frac=1)
df_0_sam
# df_1_sam
```

# make class = 0 same number of rows

```python
df_0_sam_short = df_0_sam[:df_1_sam.shape[0]]
df_0_sam_short
```

```python
df_concat = pd.concat([df_0_sam_short, df_1_sam],axis=0)
df_concat = df_concat.sample(frac = 1)
```
