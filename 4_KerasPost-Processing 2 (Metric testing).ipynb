# Putting it all tgt

```python
import os
import pandas as pd
for i in os.listdir('./results/')[:-1]:
    print(i)
    correct_count = 0
    count_ones = 0
    df = pd.read_csv('./results/' + i,index_col = 0)
    for i in range(len(df)):
        if df.loc[i, 'class'] == 1:

            if df.loc[i, 'binary_pred'] == 1:
                correct_count+=1
            count_ones+=1
    print("correct_count:",correct_count)
    print("count_ones:",count_ones)
    print('Score: {}'.format(correct_count/count_ones))
    print('\n')
```
